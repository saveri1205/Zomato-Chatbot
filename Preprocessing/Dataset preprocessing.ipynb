```python
#Unique locations in the dataset
num_unique_locations=len(data['location'].unique())
unique_locations=list(data['location'].unique())
print(unique_locations,num_unique_locations)
```

```python
final_list=[]
```

```python
#format of expected training set 
#         {"tag": "greeting",
#          "patterns": ["Hi there", "How are you", "Is anyone there?","Hey","Hola", "Hello", "Good day"],
#          "responses": ["Hello, thanks for asking", "Good to see you again", "Hi there, how can I help?"],
#          "context": [""]}

# Location wise top 5 restaurants, location wise new restaurants
l=[]
for i in unique_locations:
    if(type(i)==str):
        new_entry=dict()
        new_entry["tag"]=i
        new_entry["patterns"]=["Which are the best restaurants in "+i+" ?","Top restaurants in "+i,i+"\'s best restaurants"]
        restaurant_list=set(data[data['location']==i]['name'].head())
        if(len(restaurant_list)>0):
            restaurants=" , ".join(restaurant_list)
            new_entry["responses"]=["The best restaurants in "+i+" are "+restaurants]
        else:
            new_entry["responses"]=["Sorry, no restaurants available in this area."]
        new_entry["context"]=[""]
        l.append(new_entry)
        
        new_entry=dict()
        new_entry["tag"]="new restaurant in "+i
        new_entry["patterns"]=["Any new restaurants in "+ i +" ?"]
        new_restaurants=list((data[ (data['rate']=="NEW") & (data['location']==i)]['name']).head(10))
        if(len(new_restaurants)>0):
            new_restaurants =", ".join(new_restaurants)
            new_entry["responses"]=["New restaurants in "+i+" are "+new_restaurants]
        else:
            new_entry["responses"]=["Sorry, no new restaurants available in this area."]
        new_entry["context"]=[""]
        l.append(new_entry)
        
print(l)
final_list.extend(l)
```

```python
#restaurant types - eg Casual Dining, Food Courts etc. in Bangalore
num_unique_types=len(data['rest_type'].unique())
unique_types=list(data['rest_type'].unique())

l=[]
for i in unique_types:
    if(type(i)==str):
        new_entry=dict()
        new_entry["tag"]=i
        new_entry["patterns"]=["Best "+i+"'s in Bangalore","Which are the best "+i+"\'s in Bangalore?"]
        x=set(data[data['rest_type']==i]['name'].head())
        if(len(x)>0):
            restaurants=" , ".join(x)
            new_entry["responses"]=["The best "+i+"\'s in Bangalore are "+restaurants]
        else:
            new_entry["responses"]=["Sorry, no "+i+"'s available."]
        new_entry["context"]=[""]
        l.append(new_entry)
print(l)
final_list.extend(l)
```

```python
#converting cost values to int from string
unique_costs=list(data['approx_cost(for two people)'].unique())
mapping = dict()
for i in unique_costs:
        if i not in mapping:
            if(type(i)==str):
                mapping[i]=int(i.replace(",",""))
data.replace({'approx_cost(for two people)': mapping},inplace=True)
```

```python
# cuisines - chosen some popular cuisines
num_unique_cuisine=len(data['cuisines'].unique())
unique_cuisines=list(data['cuisines'].unique())
completely_unique_cuisines=[]
for i in unique_cuisines:
    if(type(i)==str):
        completely_unique_cuisines.extend(i.split(','))
completely_unique_cuisines=list(set(completely_unique_cuisines)) 

chosen_cuisines = [' Rolls', 'Momos','Tea',' Andhra',' Coffee',' Kashmiri','BBQ',' Bakery','Fast Food','Rajasthani',
 'Chettinad','Burger',' Ice Cream',' North Indian',' Kerala',' Cafe']
l=[]
for i in chosen_cuisines:
    for j in unique_locations:
        if(type(i)==str and type(j)==str):
            new_entry=dict()
            new_entry["tag"]=i+" "+j
            new_entry["patterns"]=["Best place to get "+i+" in "+j,"Where can I get the best "+i+" in "+j+" ?"]
            data1=data[data['location']==j]
            restaurants=list(data1['name'].head(10))
            if(len(restaurants)>0):
                restaurants =", ".join(restaurants)
                new_entry["responses"]=["The best "+i+"\'s in ",j," are available at "+restaurants]
            else:
                new_entry["responses"]=["Sorry, this cuisine is unavailable!"]
            new_entry["context"]=[""]
            l.append(new_entry)
print(l)
final_list.extend(l)

#cost wise in location questions
l=[]
#data.replace({'set': mapping, 'tesst': mapping})
chosen_costs=[500,1000]
for i in unique_locations:
    if(type(i)==str):
        data1=data[data['location']==i]
        cost_500=list(set(data1[data1["approx_cost(for two people)"]<=500]['name']))
        cost_1000=list(set(data1[(data1['approx_cost(for two people)']<=1000) & (data1['approx_cost(for two people)']>500)]))
        cost_gt1000=list(set(data1[data1['approx_cost(for two people)']>1000]))
        new_entry=dict()
        new_entry["tag"]=i+" "+"<500"
        new_entry["patterns"]=["Places in "+i+" with cost approximately Rs. 500 for two ","Which restaurants in "+i+" have prices near 500 for two people?"]
        if(len(cost_500)>0):
            restaurants =", ".join(cost_500)
            new_entry["responses"]=["Restaurants in "+i+" with cost less than 500 are "+restaurants]
        else:
            new_entry["responses"]=["Sorry, no restaurants in this area within the selected price range!"]
        new_entry["context"]=[""]
        l.append(new_entry)

        new_entry=dict()
        new_entry["tag"]=i+" "+"<1000"
        new_entry["patterns"]=["Places in "+i+" with cost approximately Rs. 1000 for two","Which restaurants in "+i+" have prices near 1000 for two people?"]
        if(len(cost_1000)>0):
            restaurants =", ".join(cost_1000)
            new_entry["responses"]=["Restaurants in "+i+" with cost less than 500 are "+restaurants]
        else:
            new_entry["responses"]=["Sorry, no restaurants in this area within the selected price range!"]
        new_entry["context"]=[""]
        l.append(new_entry)

        new_entry=dict()
        new_entry["tag"]=i+" "+">1000"
        new_entry["patterns"]=["Places in "+i+" with cost approximately Rs. 1500 and above for two ","Which restaurants in "+i+" have prices near 1500 or more for two people?"]
        if(len(cost_gt1000)>0):
            restaurants =", ".join(cost_gt1000)
            new_entry["responses"]=["Restaurants in "+i+" with cost less than 500 are "+restaurants]
        else:
            new_entry["responses"]=["Sorry, no restaurants in this area within the selected price range!"]
        new_entry["context"]=[""]
        l.append(new_entry)
print(l)
final_list.extend(l)
```

```python
num_unique_restaurants=len(data['name'].unique())
unique_restaurants=list(data['name'].unique())
num_unique_restaurants
```

```python
#branch wise restaurant details - address, phone, rating
l=[]
for i in unique_restaurants:
    new_entry=dict()
    new_entry["tag"]=i+"_locations"
    new_entry["patterns"]=["Give me the branches of "+i, "Where is "+i+" located?","Where can I find the restuarant "+i+" ?"]
    data1=data[data['name']==i]
    data1.head()
    branches=list(data1['location'])
    branches=list(set(branches))
    branches=[str(i) for i in branches]
    branches_str =", ".join(branches)
    if(len(branches)>0):
        new_entry["responses"]=["Available branches of "+i+" are: "+branches_str]
    else:
        new_entry["responses"]=["Sorry, no branches of this restaurant are open."]
    new_entry["context"]=[""]
    l.append(new_entry)
    for j in branches:
        new_entry=dict()
        new_entry["tag"]=i+"_"+j+"_address"
        new_entry["patterns"] = ["What's the address of "+i+", "+j+" ?","Give me address of "+i+", "+j]
        new_entry["context"]=[""]
        data2=list(data1[data1['location']==j]['address'])
        if(len(data2)>0):
            address = data2[0]
            if(address):
                new_entry["response"] = ["The address of "+i+", "+j+" is : "+address]
            else:
                new_entry["response"] = ["Sorry, we do not have this information!","This information is unavailable, please try another question."]
            l.append(new_entry)

        new_entry=dict()
        new_entry["tag"]=i+"_"+j+"_contact"
        new_entry["context"]=[""]
        new_entry["patterns"] = ["What's the contact number of "+i+", "+j+" ?","Give me the phone number of "+i+", "+j]
        data2=list(data1[data1['location']==j]['phone'])
        if(len(data2)>0):
            phone = data2[0]
            if phone:
                new_entry["response"] = ["The contact number of "+i+", "+j+" is : "+str(phone)]
            else:
              new_entry["response"] = ["Sorry, we do not have this information!","This information is unavailable, please try another question."]
            l.append(new_entry)

        new_entry=dict()
        new_entry["tag"]=i+"_"+j+"_rating"
        new_entry["context"]=[""]
        new_entry["patterns"] = ["What's the rating of "+i+", "+j+" ?","Give me the rating of "+i+", "+j]
        data2=list(data1[data1['location']==j]['rate'])
        if(len(data2)>0):
            rate = data2[0]
            if(rate):
                new_entry["response"] = ["The rating of "+i+", "+j+" is : "+str(rate)]
            else:
                new_entry["response"] = ["Sorry, we do not have this information!","This information is unavailable, please try another question."]
            l.append(new_entry)

#print(l)
final_list.extend(l)
```
